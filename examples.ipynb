# **INFO PAIS**

```python
df_pais = pd.read_csv(r'/content/drive/MyDrive/Colab Notebooks/Master Python DS-ML/data/Info_pais.csv', encoding = "ISO-8859-1", delimiter=';')
df_pais.head()
```

```python
df_pais_order = df_pais.sort_values("Esperanza de vida", ascending=True)
df_pais_order.head()
```

```python
plt.scatter(df_pais_order["Renta per capita"], df_pais_order["Esperanza de vida"])  # Si no lo muestra ejecutar plt.show() a continuación
plt.title("Renta per capita & Esperanza")
plt.xlabel("Renta per capita")
plt.ylabel("Espezanza de vida")
```

```python
# Queremos una nueva columna donde normalice la población
df_pais_order["Pobl_norm"] = df_pais_order["Poblacion"]/(max(df_pais_order["Poblacion"]) / 10000) # Se puede dividir el max entre 10000 para que no salgan numeros muy pequeños
df_pais_order.head()
```

```python
plt.scatter(df_pais_order["Renta per capita"], df_pais_order["Esperanza de vida"], s=df_pais_order["Pobl_norm"], c=df_pais_order["Pobl_norm"])  # Si no lo muestra ejecutar plt.show() a continuación
plt.title("Renta per capita & Esperanza")
plt.xlabel("Renta per capita")
plt.ylabel("Espezanza de vida")
```

```python
# Ampliamos imagen
fig = plt.gcf()
fig.set_size_inches(14.5, 10)

# Etiquetamos los 10 primeros paises
for i in range(1, 10):
  plt.annotate(df_pais_order["País"][i], (df_pais_order["Renta per capita"][i], df_pais_order["Esperanza de vida"][i]))
```
